```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart_disease_dataset.csv') 
df.head()
```

```python
def discretize_heart_data(df):
    # Create a copy to avoid SettingWithCopy warnings
    df_bin = df.copy()

    # --- 1. CONTINUOUS VARIABLES (Binning) ---

    # Age (Standard risk groups)
    df_bin['Age_Bin'] = pd.cut(df_bin['age'], 
                               bins=[0, 45, 60, 120], 
                               labels=['Young', 'Middle', 'Old'])

    # Trestbps (Blood Pressure) - Based on PDF Cite [21] (140 threshold)
    df_bin['BP_Bin'] = pd.cut(df_bin['trestbps'], 
                              bins=[0, 120, 140, 300], 
                              labels=['Normal', 'Elevated', 'High_BP'], 
                              right=False) # right=False means 140 falls into 'High_BP'

    # Chol (Cholesterol) - Based on PDF Cite [25] (200/240 thresholds)
    df_bin['Chol_Bin'] = pd.cut(df_bin['chol'], 
                                bins=[0, 200, 240, 600], 
                                labels=['Desirable', 'Borderline', 'High_Chol'],
                                right=False)

    # Thalach (Max Heart Rate) - Quantiles based on PDF range [38]
    df_bin['HR_Bin'] = pd.cut(df_bin['thalach'], 
                              bins=[0, 110, 150, 250], 
                              labels=['Low_Rate', 'Normal_Rate', 'High_Rate'])

    # Oldpeak (ST Depression) - Based on PDF Cite [49] (Severity)
    # Note: We treat 0 explicitly as 'None'
    df_bin['Oldpeak_Bin'] = pd.cut(df_bin['oldpeak'], 
                                   bins=[-1, 0, 2.0, 10], 
                                   labels=['No_Depression', 'Ischemia', 'Severe_Ischemia'])

    # --- 2. CATEGORICAL VARIABLES (Mapping Labels) ---
    
    # Sex [Cite: 12]
    df_bin['Sex_Label'] = df_bin['sex'].map({1: 'Male', 0: 'Female'})

    #ca
    df_bin['CA_Label'] = df_bin['ca'].astype(str) + "_Vessels"

    # Chest Pain (cp) [Cite: 16]
    df_bin['CP_Label'] = df_bin['cp'].map({
        1: 'Typical_Angina',
        2: 'Atypical_Angina',
        3: 'Non_Anginal',
        4: 'Asymptomatic'
    })

    # Fasting Blood Sugar (fbs) [Cite: 28]
    df_bin['FBS_Label'] = df_bin['fbs'].map({1: 'High_Sugar', 0: 'Normal_Sugar'})

    # RestECG [Cite: 32]
    df_bin['ECG_Label'] = df_bin['restecg'].map({
        0: 'Normal',
        1: 'ST_Abnorm',
        2: 'LVH'
    })

    # Exang [Cite: 42]
    df_bin['Exang_Label'] = df_bin['exang'].map({1: 'Yes', 0: 'No'})

    # Slope [Cite: 51]
    df_bin['Slope_Label'] = df_bin['slope'].map({
        1: 'Upsloping',
        2: 'Flat',
        3: 'Downsloping'
    })

    # Thal [Cite: 60]
    df_bin['Thal_Label'] = df_bin['thal'].map({
        3: 'Normal',
        6: 'Fixed_Defect',
        7: 'Reversible_Defect'
    })

    # Num (Target) - Grouping 1-4 as Positive [Cite: 69]
    df_bin['Disease_Target'] = df_bin['num'].apply(lambda x: 'Positive' if x > 0 else 'Negative')

    # Return only the new columns needed for the Bayesian Network
    cols_to_keep = ['Age_Bin', 'Sex_Label', 'CP_Label', 'BP_Bin', 'Chol_Bin', 
                    'FBS_Label', 'ECG_Label', 'HR_Bin', 'Exang_Label', 
                    'Oldpeak_Bin', 'Slope_Label', 'Thal_Label', 'CA_Label', 'Disease_Target']
    
    return df_bin[cols_to_keep]
```

```python
formatted_df = discretize_heart_data(df)
print(formatted_df.head())
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import BayesianEstimator
import itertools
```

```python
model = DiscreteBayesianNetwork([
    # --- Tier 1 (Roots) to Tier 2 (Clinical) ---
    ('Sex_Label', 'BP_Bin'),
    ('Sex_Label', 'Chol_Bin'),
    ('Age_Bin', 'BP_Bin'),
    ('Age_Bin', 'Chol_Bin'),

    # --- Tier 1 (Roots) to Tier 3 (Exercise/Other) ---
    ('Age_Bin', 'HR_Bin'),
    ('Age_Bin', 'CA_Label'),

    # --- Tier 2 (Clinical) to Tier 3 (Exercise/Other) ---
    ('Chol_Bin', 'CA_Label'),

    # --- Tier 3 Inter-connections ---
    ('HR_Bin', 'CA_Label'),

    # --- Direct Parents to Tier 4 (Target Disease/num) ---
    ('Sex_Label', 'Disease_Target'),
    ('BP_Bin', 'Disease_Target'),
    ('Chol_Bin', 'Disease_Target'),
    ('HR_Bin', 'Disease_Target'),
    ('Oldpeak_Bin', 'Disease_Target'),
    ('Slope_Label', 'Disease_Target'),
    ('CA_Label', 'Disease_Target'),
    ('Thal_Label', 'Disease_Target'),
    ('FBS_Label', 'Disease_Target'),
    ('ECG_Label', 'Disease_Target'),      # <--- UPDATED to match your function

    # --- Tier 4 (Target) to Tier 5 (Symptoms) ---
    ('Disease_Target', 'Exang_Label'),
    ('Disease_Target', 'CP_Label')
])
```

```python
model.fit(formatted_df, estimator=BayesianEstimator, prior_type='BDeu', equivalent_sample_size=10)
```

```python
cpd_target = model.get_cpds('Disease_Target')
```

```python
def get_formatted_cpt_table(cpd):
    # Get all variables involved
    target_var = cpd.variable
    parent_vars = cpd.variables[1:] # pgmpy stores [target, parent1, parent2...]
    
    # Get state names for every variable
    state_names = [cpd.state_names[var] for var in cpd.variables]
    
    # Create every combination of states
    combinations = list(itertools.product(*state_names))
    values = cpd.values.flatten() # Flatten the probability array
    
    rows = []
    for i, combo in enumerate(combinations):
        # combo[0] is the Target State, the rest are Parents
        row = {
            'Target': combo[0],
            'Probability': f"{values[i]:.4f}"
        }
        
        # Add a column for each parent variable dynamically
        for parent, state in zip(parent_vars, combo[1:]):
            row[parent] = state
            
        rows.append(row)
        
    return pd.DataFrame(rows)
```

```python
readable_cpt = get_formatted_cpt_table(cpd_target)
```

```python
cols = [c for c in readable_cpt.columns if c not in ['Target', 'Probability']] + ['Target', 'Probability']
readable_cpt = readable_cpt[cols]
```

```python
from pgmpy.inference import VariableElimination
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Prepare the Inference Engine
infer = VariableElimination(model)

# 2. Define a function to predict 'num' for a single row
def predict_heart_disease(row):
    # Prepare evidence dictionary
    # We must exclude the target ('Disease_Target') and any columns not in the model
    evidence = {
        'Age_Bin': row['Age_Bin'],
        'Sex_Label': row['Sex_Label'],
        'BP_Bin': row['BP_Bin'],
        'Chol_Bin': row['Chol_Bin'],
        'HR_Bin': row['HR_Bin'],
        'Oldpeak_Bin': row['Oldpeak_Bin'],
        'Slope_Label': row['Slope_Label'],
        'CA_Label': row['CA_Label'],
        'Thal_Label': row['Thal_Label']
        # Add 'FBS_Label' here if you included it in the model structure
    }
    
    try:
        # Query the network
        # We ask: What is the most likely state for 'Disease_Target'?
        q = infer.map_query(variables=['Disease_Target'], evidence=evidence, show_progress=False)
        return q['Disease_Target']
    except:
        return "Error"

print("Starting predictions (this might take a minute)...")

# 3. Run prediction on the entire dataset (Testing against sample data)
# Note: In a real ML project, you would split Train/Test. 
# For this BN exercise, testing on the full set is often accepted to show CPT consistency.
formatted_df['Predicted_Target'] = formatted_df.apply(predict_heart_disease, axis=1)

# 4. Calculate Accuracy
actual = formatted_df['Disease_Target']
predicted = formatted_df['Predicted_Target']

acc = accuracy_score(actual, predicted)
print(f"\nModel Accuracy: {acc:.2%}")

# 5. Generate Confusion Matrix
cm = confusion_matrix(actual, predicted, labels=['Negative', 'Positive'])

# Display Text Report
print("\nClassification Report:")
print(classification_report(actual, predicted, labels=['Negative', 'Positive']))

# Optional: Plot the matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Pred: No Disease', 'Pred: Disease'],
            yticklabels=['Actual: No Disease', 'Actual: Disease'])
plt.title('Bayesian Network Confusion Matrix')
plt.show()
```

### Conditional Probability Table: Heart Rate given Age
This represents the distribution $P(\text{Heart Rate} \mid \text{Age})$.

**1. Given Age = Young:**
* $P(\text{High\_Rate} \mid \text{Young}) = 0.7253$
* $P(\text{Normal\_Rate} \mid \text{Young}) = 0.2580$
* $P(\text{Low\_Rate} \mid \text{Young}) = 0.0168$

**2. Given Age = Middle:**
* $P(\text{High\_Rate} \mid \text{Middle}) = 0.5544$
* $P(\text{Normal\_Rate} \mid \text{Middle}) = 0.3840$
* $P(\text{Low\_Rate} \mid \text{Middle}) = 0.0615$

**3. Given Age = Old:**
* $P(\text{High\_Rate} \mid \text{Old}) = 0.3414$
* $P(\text{Normal\_Rate} \mid \text{Old}) = 0.5358$
* $P(\text{Low\_Rate} \mid \text{Old}) = 0.1228$

```python
from pgmpy.inference import VariableElimination

infer = VariableElimination(model)

ages = ['Young', 'Middle', 'Old']

results_age = []
for age in ages:
    # Query: What is the prob of Disease given specific Age?
    q = infer.query(variables=['Disease_Target'], evidence={'Age_Bin': age})
    
    # Extract the probability of "Positive" disease (usually index 1)
    # We get the value corresponding to 'Positive'
    prob_disease = q.values[q.name_to_no['Disease_Target']['Positive']]
    
    results_age.append({
        'Age Group': age,
        'Probability of Disease': prob_disease
    })


age_disease_df = pd.DataFrame(results_age)
print("--- Relationship: Age -> Heart Disease Risk ---")
# Format as percentage for better readability
age_disease_df['Probability of Disease'] = age_disease_df['Probability of Disease']
display(age_disease_df)
```

This CPT represents:
* $P(\text{Disease} \mid \text{Young})$
* $P(\text{Disease} \mid \text{Middle})$
* $P(\text{Disease} \mid \text{Old})$

```python
from IPython.display import display
import pandas as pd

def analyze_connection(parent, child, model, infer):
    """
    Calculates the probability of the Child node states given the Parent node states.
    """
    print(f"\n--- Connection: {parent} -> {child} ---")
    
    # Get all possible states for the parent variable
    parent_states = model.get_cpds(parent).state_names[parent]
    
    # Store results
    results = []
    
    for state in parent_states:
        # Query the inference engine
        q = infer.query(variables=[child], evidence={parent: state}, show_progress=False)
        
        # Get probabilities and state names for the child
        child_values = q.values
        child_states = q.state_names[child]
        
        # Create a row for this parent state
        row = {parent: state}
        for c_state, prob in zip(child_states, child_values):
            row[f"Prob({c_state})"] = float(f"{prob:.4f}")
            
        results.append(row)
    
    # Create and display DataFrame
    df_result = pd.DataFrame(results)
    
    # Optional: Highlight high probabilities
    # df_result = df_result.style.background_gradient(cmap='Blues', axis=1)
    
    display(df_result)

# Initialize inference engine if not already active
infer = VariableElimination(model)
```
